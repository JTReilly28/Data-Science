## üìç **Section 2: DataFrame - Two-Dimensional Data**

‚úÖ 1. Tabular Data Representation
Mimics a spreadsheet or SQL table: rows + labeled columns.

Makes data easy to read, access, and interpret.

‚úÖ 2. Labeled Indexing
You can access data using row/column labels, not just numbers.

This improves clarity and reduces errors in complex analyses.

```python
# Create a datetime index
dates = pd.date_range("20220101", periods=6)
print(dates)
# Create a DataFrame with random numbers
df = pd.DataFrame(np.random.randn(6, 4), index=dates, columns=list("ABCD"))
df
```

### üìò Explanation:
* `date_range()` creates dates.
* `np.random.randn()` creates random data.
* `DataFrame(...)` builds a table from the array.

Use what you've learned to:
* Create a DataFrame with 5 students and their grades in Math, Science, and English.
* Calculate the average for each student.
* Sort the DataFrame by average grade.

```python
final = pd.DataFrame({
    "Student": ["J.T.", "Ethan", "Hong", "Tony", "Bob"],
    "Math Grade": [98, 42, 77, 90, 68],
    "Science Grade": [95, 39, 88, 75, 96],
    "English Grade": [87, 88, 33, 56, 97],
})
final["Average"] = (final["Math Grade"] + final["Science Grade"] + final["English Grade"]) / 3
sorted_final = final.sort_values(by="Average")
print(sorted_final)
print(final.describe())
```
